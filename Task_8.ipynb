##6.  Write a Python programming to display a bar chart of the popularity of programming Languages.
```
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7
The code snippet gives the output shown in the following screenshot:
```

image.png

````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# define x and y values

Programming_languages= "Java", "Python", "PHP", "JavaScript", "C"," C++"
Popularity = 22.2, 17.6, 8.8, 8, 7.7, 6.7

# bar is used for creating bar graph in matplotlib
plt.bar(Programming_languages,Popularity )

# title
plt.title("Popularity of Programming language\nworldwide,Oct 2017 comapred to a year ago")

#labels x and y
plt.xlabel("javaScript\nlanguages")
plt.ylabel("Popularity")

# Grid
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.minorticks_on()
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='red')


# show the graph
plt.show()
````

##Exercise 5: Read face cream and facewash product sales data and show it using the bar chart
The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

The bar chart should look like this.

image.png

````python

import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv('company_sales_data.csv')

# Extract face cream and facewash product sales data and month number
month_number = data['month_number']
facecream = data['facecream']
facewash = data['facewash']

# Set size
plt.figure(figsize=(8, 6))  # Adjust figure size (width, height)

# Plotting the bar chart
# month_number + 0.15: Adjusts the x-position of bars for Face Wash sales data. This shifts the bars to the right by 0.15 units.
# month_number - 0.15: Adjusts the x-position of bars for Face Cream sales data. This shifts the bars to the left by 0.15 units.
plt.bar(month_number - 0.15, facecream, width=0.30, color='blue', label='Face Cream sales data')
plt.bar(month_number + 0.15, facewash, width=0.30, color='orange', label='Face Wash sales data')

# Set the x-axis tick positions to match the month_number
plt.xticks(month_number)

# Add labels and title
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title('Facewash and facecream sales data')

# dashed line grid
plt.grid(True,color='#B0B0B0', linestyle='--', linewidth=1)

# Show legend at upper left location
plt.legend(loc='upper left')

# Adjust layout for better appearance
plt.tight_layout()

# Display the plot
plt.show()
````
